## Chaves na Mão

```python
chavesnamao = data['chavesnamao']
```

```python
df_chavesnamao = pd.DataFrame.from_dict(chavesnamao, orient='index')
```

```python
# Avaliar os dados
df_chavesnamao.head()
```

```python
# Verificar o formato dos dados
df_chavesnamao.info()
```

```python
# Avaliar valores nulos
df_chavesnamao.isnull().sum()
```

```python
# Transformar data e hora em datetime
df_chavesnamao['data_extracao'] = pd.to_datetime(df_chavesnamao['data_extracao'], format='%Y-%m-%d %H:%M:%S.%f')
```

```python
# Retirar o milisegundos
df_chavesnamao['data_extracao'] = df_chavesnamao['data_extracao'].dt.floor('S')

# Criar coluna apenas para a data da extração
df_chavesnamao['Data da Extração'] = df_chavesnamao['data_extracao'].dt.date

# Criar a coluna apenas para hora da extração
df_chavesnamao['Hora da Extração'] = df_chavesnamao['data_extracao'].dt.time

# Remover a coluna data_extracao
df_chavesnamao = df_chavesnamao.drop('data_extracao', axis=1)
```

```python
print(df_chavesnamao['Hora da Extração'].dtype)
```

```python
# Retirar R$ da coluna ALUGUEL
df_chavesnamao['ALUGUEL'] = df_chavesnamao['ALUGUEL'].str.replace('R$ ', '')

# Retirar o ponto dos valores
df_chavesnamao['ALUGUEL'] = df_chavesnamao['ALUGUEL'].str.replace('.', '')

# Transformar os valores str para float
df_chavesnamao['ALUGUEL'] = df_chavesnamao['ALUGUEL'].astype(float)
```

```python
# Retirar R$ da coluna CONDOMINIO
df_chavesnamao['CONDOMINIO'] = df_chavesnamao['CONDOMINIO'].str.replace('R$ ', '')

# Retirar /mês da coluna CONDOMINIO
df_chavesnamao['CONDOMINIO'] = df_chavesnamao['CONDOMINIO'].str.replace('/mês', '')

# Inserir 0 no lugar de "Não informado"
df_chavesnamao['CONDOMINIO'] = df_chavesnamao['CONDOMINIO'].str.replace('Não informado', '0')

# Converter str para float da coluna CONDOMINIO
df_chavesnamao['CONDOMINIO'] = df_chavesnamao['CONDOMINIO'].astype(float)
```

```python
# Retirar R$ da coluna IPTU
df_chavesnamao['IPTU'] = df_chavesnamao['IPTU'].str.replace('R$ ', '')

# Retirar /mês da coluna IPTU
df_chavesnamao['IPTU'] = df_chavesnamao['IPTU'].str.replace('/mês', '')

# Inserir 0 no lugar de "Não informado"
df_chavesnamao['IPTU'] = df_chavesnamao['IPTU'].str.replace('Não informado', '0')

# Converter str para float da coluna IPTU
df_chavesnamao['IPTU'] = df_chavesnamao['IPTU'].astype(float)
```

```python
# Retirar R$ da coluna ALUGUEL_COND
df_chavesnamao['ALUGUEL_COND'] = df_chavesnamao['ALUGUEL_COND'].str.replace('R$ ', '')

# Retirar /mês da coluna ALUGUEL_COND
df_chavesnamao['ALUGUEL_COND'] = df_chavesnamao['ALUGUEL_COND'].str.replace('/mês', '')

# Inserir 0 no lugar de "N/A"
df_chavesnamao['ALUGUEL_COND'] = df_chavesnamao['ALUGUEL_COND'].str.replace('N/A', '0')

# Converter str para float
df_chavesnamao['ALUGUEL_COND'] = df_chavesnamao['ALUGUEL_COND'].astype(float)
```

```python
df_chavesnamao.head()
```

```python
# Renomear a coluna INFO_1 para Área Útil (m²)
df_chavesnamao.rename(columns={'INFO_1': 'Área útil/total (m²)'}, inplace=True)

# Retira Area útil/total  \n da informação da coluna Área útil tal como o m²
df_chavesnamao['Área útil/total (m²)'] = df_chavesnamao['Área útil/total (m²)'].str.replace('Area útil/total  \n|Area útil  \n|Area total  \n', '', regex=True)

# Retirar m² da coluna
df_chavesnamao['Área útil/total (m²)'] = df_chavesnamao['Área útil/total (m²)'].str.replace('m²', '')

# Dividir a área útil da área total a partir do delimitador /
df_chavesnamao[['Área Útil (m²)', 'Área Total (m²)']] = df_chavesnamao['Área útil/total (m²)'].str.split('/', expand=True)

# Substituir None para 0 na coluna Área Total (m²)
df_chavesnamao['Área Total (m²)'] = df_chavesnamao['Área Total (m²)'].fillna(0)

#Converter a coluna Área últil (m²) em númerico
df_chavesnamao['Área Útil (m²)'] = df_chavesnamao['Área Útil (m²)'].astype(int)

#Converter a coluna Área Total (m²) em númerico
df_chavesnamao['Área Total (m²)'] = df_chavesnamao['Área Total (m²)'].astype(int)
```
